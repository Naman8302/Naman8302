```python
from wordcloud import WordCloud
text_question = ' '.join(df_train[df_train['label'] == 1]['question'])
wordcloud_questions = WordCloud(width=800,height=400,background_color = 'white').generate(text_question)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud_questions,interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of AI Generated Questions')
plt.show()
```
